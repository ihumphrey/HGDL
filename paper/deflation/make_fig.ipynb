```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})

x0 = np.zeros(1)
alpha = 1
r = 1
deflation = lambda x: (np.exp(-alpha/(r**2-np.linalg.norm(x-x0)**2) + alpha/r**2) if np.linalg.norm(x-x0)<r else 0)
x = np.arange(-1.5, 1.5, 1e-5)
y = np.array([deflation(z) for z in x])


fig, ax = plt.subplots(4, 1, figsize=(8.3,11.7))
plt.subplots_adjust(hspace=0.6)

plt.subplot(411)
plt.title('Bump Function')
plt.plot(x, y, color='black')
plt.xticks([-1, 0, 1], [r'$x_0-r$', r'$x_0$', r'$x_0+r$'])

plt.subplot(412)
plt.yscale('log')
plt.plot(x, 1./(1-y), color='black')
plt.xticks([-1, 0, 1], [r'$x_0-r$', r'$x_0$', r'$x_0+r$'])
plt.yticks(10**np.arange(9)[::2])
plt.title(r"$\frac{1}{1-bump}$")

plt.subplot(413)
plt.plot(x, x**2+1, color='black', label=r'$y=x^2+1$')
plt.plot(x, 2*x, color='black', ls='--', label=r"$y'=2x$")
plt.xticks([-1, 0, 1], [r'$x_0-r$', r'$x_0$', r'$x_0+r$'])
plt.yticks([])
plt.legend()
plt.title('Simple Function and Gradient')
#"""
plt.subplot(414)
plt.plot(x, 2*x/(1-y), color='black')
plt.xticks([0], [r'$x_0$'])
#plt.yticks(10**np.arange(9)[::2])
#plt.ylim(-1,1)
#plt.xticks([-1e-2, 0, 1e-2], [r'$x_0-\epsilon$', r'$x_0$', r'$x_0+\epsilon$'])
plt.xticks([-1, 0, 1], [r'$x_0-r$', r'$x_0$', r'$x_0+r$'])
plt.yscale('symlog')
plt.yticks(list(10**np.arange(0,5)[::2]) + list((-1)*10**np.arange(5)[::2][1:]))
#plt.xlim(-2e-2, 2e-2)
plt.title("Deflated Gradient")
#"""
plt.savefig('bump.png')
```
