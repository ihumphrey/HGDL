```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})

x0 = np.zeros(1)
alpha = 1
r = 1
deflation = lambda x: (np.exp(-alpha/(r**2-np.linalg.norm(x-x0)**2) + alpha/r**2) if np.linalg.norm(x-x0)<r else 0)
x = np.arange(-1.5, 1.5, 1e-4)
y = np.array([deflation(z) for z in x])


fig, ax = plt.subplots(2, 1, figsize=(8.3,11.7/2))
plt.subplots_adjust(hspace=0.6)
plt.subplot(211)

plt.title('Bump Function')
plt.plot(x, y, color='black')
plt.xticks([0], [r'$x_0$'])
plt.xticks([0,1])

plt.subplot(212)
plt.yscale('log')
plt.plot(x, 1./(1-y), color='black')
plt.xticks([0], [r'$x_0$'])
plt.yticks(10**np.arange(9)[::2])
plt.title(r"$\frac{1}{1-bump}$")

plt.savefig('bump.png')
```
